```python
# Import lớp Image
from PIL import Image
```

```python
# Thư mục chứa hình ảnh
dir = 'C:/Users'
```

```python
# Đường dẫn đến hình ảnh
image_path = dir +'/Go toc do.png'
print(image_path)
```

```python
# Sử dụng Image.open() đọc ảnh
img01 = Image.open(image_path)
```

```python
# Hiển thị hình ảnh
display(img01)
```

```python
# In một số thông tin
# In định dạng ảnh
print("Định dang ảnh: ", img01.format)
```

```python
# In ra size
print("Kích thước hình ảnh: ", img01.size)
```

```python
# Chuyển đổi hình ảnh từ RGBA sang RGB
img01_rgb = img01.convert("RGB")

# Lưu ảnh dưới dạng JPG và PNG
new_img_01_path = 'C:/Users/Admin/Documents/Go_toc_do.JPG'
new_img_png_01_path = 'C:/Users/Admin/Documents/Go_toc_do.PNG'

img01_rgb.save(new_img_01_path)  # Lưu dưới dạng JPEG
img01.save(new_img_png_01_path)  # PNG hỗ trợ RGBA, nên không cần chuyển đổi
```

```python
# Đóng tập tin
img01.close()
```
