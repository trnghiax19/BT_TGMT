```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 20, 25, 30]

x1 = [1, 2, 3, 4, 5]
y1 = [20, 15, 8, 9, 20]
     
plt.plot(x,y, 'bo-', label = 'Dữ liệu')
plt.plot(x1,y1, marker='o', color='r', linestyle='-', label='Dữ liệu-1')
plt.legend()
plt.xlabel('Trục x')
plt.ylabel('Trục y')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [15, 10, 25, 12, 18]

plt.bar(categories, values, color = 'g', alpha = 0.6)
plt.xlabel('Danh mục')
plt.ylabel('Giá trị')
plt.show()
```

```python
my_labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
my_colors = ['red', 'green', 'blue', 'yellow']

plt.pie(sizes, labels=my_labels, colors = my_colors, autopct='%1.1f%%') # Vẽ biểu đồ hình tròn
plt.title('Biểu đồ hình tròn')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 16, 20, 30, 25]

plt.scatter(x, y, c='r', marker='o', label='Dữ liệu')  # Vẽ biểu đồ phân tán
plt.title('Biểu đồ phân tán ví dụ')  # Đặt tiêu đề cho biểu đồ
plt.xlabel('Trục x')  # Đặt nhãn cho trục x
plt.ylabel('Trục y')  # Đặt nhãn cho trục y
plt.legend()  # Hiển thị chú giải
plt.show()  # Hiển thị biểu đồ
```

```python
import matplotlib.pyplot as plt

data = [15, 18, 22, 30, 35, 45, 50, 55, 65]

plt.boxplot(data)  # Vẽ biểu đồ hộp
plt.title('Biểu đồ hộp ví dụ')  # Đặt tiêu đề cho biểu đồ
plt.ylabel('Giá trị')  # Đặt nhãn cho trục y
plt.show()  # Hiển thị biểu đồ
```
